# **Arrays**

```python
class Array:
    def __init__(self):
        self.arr = []

    # CREATE
    def create(self, elements):
        self.arr = elements.copy()

    # DISPLAY / TRAVERSE
    def display(self):
        print("Array:", self.arr)

    # INSERT AT ANY POSITION
    def insert(self, index, value):
        if index < 0 or index > len(self.arr):
            print("Invalid index!")
        else:
            self.arr.insert(index, value)

    # DELETE AT ANY POSITION
    def delete(self, index):
        if index < 0 or index >= len(self.arr):
            print("Invalid index!")
        else:
            removed = self.arr.pop(index)
            print("Deleted:", removed)

    # SEARCH (Linear Search)
    def search(self, value):
        for i in range(len(self.arr)):
            if self.arr[i] == value:
                return i
        return -1

    # UPDATE (Modify value at index)
    def update(self, index, value):
        if index < 0 or index >= len(self.arr):
            print("Invalid index!")
        else:
            self.arr[index] = value
```

```python
# Examples
A = Array()

# Creation
A.create([10, 20, 30, 40, 50])
A.display()

# Insertion
A.insert(2, 99)      # insert 99 at index 2
A.display()

# Deletion
A.delete(3)          # delete element at index 3
A.display()

# Search
idx = A.search(30)
print("30 found at index:", idx)

# Update
A.update(1, 777)
A.display()
```
