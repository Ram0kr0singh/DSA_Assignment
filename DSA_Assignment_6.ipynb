```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def insert(self, data):
        if data < self.data:
            if self.left is None:
                self.left = Node(data)
            else:
                self.left.insert(data)
        elif data > self.data:
            if self.right is None:
                self.right = Node(data)
            else:
                self.right.insert(data)

    def search(self, val):
        if val == self.data:
            return True
        if val < self.data:
            if self.left is None:
                return False
            return self.left.search(val)
        if val > self.data:
            if self.right is None:
                return False
            return self.right.search(val)

    def dfs(self):
        print(self.data, end=" ")
        if self.left:
            self.left.dfs()
        if self.right:
            self.right.dfs()

    def bfs(self):
        queue = [self]
        while len(queue) > 0:
            current = queue.pop(0)
            print(current.data, end=" ")
            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)

def main():
    print("TREE OPERATIONS")

    root = Node(10)
    root.insert(5)
    root.insert(15)
    root.insert(2)
    root.insert(7)

    print(f"Search for 7: {root.search(7)}")

    print("DFS:", end=" ")
    root.dfs()

    print("\nBFS:", end=" ")
    root.bfs()
    print()

if __name__ == "__main__":
    main()
```
